```python
import mne
import requests

# Define the URL of the EDF file
url = "https://physionet.org/files/eegmmidb/1.0.0/S001/S001R01.edf"

# Fetch the file dynamically
response = requests.get(url)
response.raise_for_status()  # Ensure the request was successful

# Save the file temporarily in Colab
edf_path = "/tmp/eeg.edf"
with open(edf_path, "wb") as f:
    f.write(response.content)

# Read the EDF file using MNE
raw = mne.io.read_raw_edf(edf_path, preload=True)

# Plot EEG data (optional)
raw.plot()
```

```python
raw.filter(l_freq=0.5, h_freq=50, fir_design='firwin')
```
